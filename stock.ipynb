```python
plt.figure(figsize=(16,9))
stock['High'].value_counts().head(8).plot.bar()
```

```python
plt.figure(figsize=(16,7))
stock['Total Trade Quantity'].value_counts().sort_index().plot.bar()
```

```python
plt.figure(2)
plt.subplot(121)
sns.distplot(stock['Last'])
plt.subplot(122)
stock['Last'].plot.box(figsize=(15,5))
```

```python
columns = ['High', 'Last', 'Total Trade Quantity','Turnover (Lacs)','Open']  # Replace with your actual column names

for column in columns:
    plt.figure(figsize=(25, 6))
    sns.countplot(x=stock[column], data=stock, palette='hls')
    plt.xticks(rotation=90)
    plt.show()
```

```python
stock.corr()
```

```python
columns = ['High', 'Last', 'Total Trade Quantity','Turnover (Lacs)']
for column in columns:
    sns.violinplot(x='Open', y=column, data=stock)
    plt.xlabel('Open')
    plt.ylabel(column)
    plt.title(f'{column} vs Open')
    plt.show()
```

```python
sns.pairplot(data=stock, hue='High', corner=True)
```

```python
plt.figure(figsize=(10,6))
sns.scatterplot(x="Total Trade Quantity", y="High", data=stock)
plt.title("Turnover (Lacs)", size=15)
plt.show()
```

```python
features = ['High', 'Last', 'Total Trade Quantity','Turnover (Lacs)']
plt.figure(figsize=(10,5))
for i in range(0,len(features)):
    plt.subplot(1, len(features), i + 1)
    sns.boxplot(y=stock[features[i]], color='CornflowerBlue', orient='v')
    plt.tight_layout()
```
